#Population growth model (Hyperbolic model)

```python
# Importing required libraries:
from scipy.integrate import odeint
import numpy as np
import matplotlib.pyplot as plt
```

image.png

```python
# Defining the given population growth model
def model(N,t):
    T0 = 2007
    g = 42
    C = 1.82e11
    dNdt = C/((T0-t)**2+g**2)
    return dNdt
```

```python
# Giving parameters of time to predict and assumed initial condition
# Prediction time is up until 2300
t = np.linspace(0,2300,500)
N = 0
result = odeint(model,N,t)
```

```python
# Seeing model results according to the given initial condition and time
result[:10]
```

```python
# Plotting result to see it graphically
plt.plot(t,result)
plt.xlabel('time')
plt.ylabel('Population size')
plt.show()
```

```python
print(f"Year {int(t[499])} has approximately population size of {result[-1]}")
```

# The differential equation has an imaginary unit. Therefore there is no 'exact' equilibrium points for the equation. It is known as "population cyles" or "population oscillations" like the predator-prey model.

```python
# The same model with varied population capacity
def model2(C,t,N):
    T0 = 2007
    g = 42
    dNdt = C/((T0-t)**2+g**2)
    return dNdt
```

```python
C = [1.82*10**11, 2*10**11, 1.50*10**11]
```

```python
result2 = odeint(model2,C,t,args=(N,))
```

```python
# Another with varieties of population capacity
fig,ax = plt.subplots()
ax.plot(t,result2[:,0],label = '1.82 Billion capacity')
ax.plot(t,result2[:,1],label = '2 Billion capacity')
ax.plot(t,result2[:,2],label = '1.5 Billion capacity')
ax.legend()
ax.set_xlabel('t')
ax.set_ylabel('Population Capacity')
```

# Although there is no equilibrium in the population growth model, it is still possible to find the slope field of the equation.

```python
# Specify the size for the slope field. Just for better readability.
# The third argument represents the space of the arrows. The bigger the number the more are the arrows!
N_vals = np.linspace(-200e6, 200e6, 20)
t_vals = np.linspace(0, 2300, 20)
N, t = np.meshgrid(N_vals,t_vals)
```

```python
# We calculate using our defined model at the start
slope_result = model(N,t)
```

```python
# quiver is a specific visualization for slope fields
plt.quiver(t,N,np.ones_like(slope_result),slope_result)
plt.xlabel('t (time)')
plt.ylabel('Population')
plt.title('Slope field')
plt.show()
```

```python
# Just in case it is hard to be seen haha
N_vals = np.linspace(-200e6, 200e6, 20)
t_vals = np.linspace(0, 2300, 100)
N, t = np.meshgrid(N_vals,t_vals)
slope_result = model(N,t)
```

```python
plt.quiver(t,N,np.ones_like(slope_result),slope_result)
plt.xlabel('t (time)')
plt.ylabel('Population')
plt.title('Slope field')
plt.show()
```
